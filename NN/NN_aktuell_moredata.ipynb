```python
import json
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
from torch.utils.data import Dataset, DataLoader, Subset
import optuna
from optuna.trial import TrialState
import pickle
import os
from datetime import datetime
import seaborn as sns

# Add after imports, before data loading
def create_output_directory(base_path="./NN"):
    """Create timestamped output directory for plots"""
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_dir = os.path.join(base_path, f"plots_{timestamp}")
    os.makedirs(output_dir, exist_ok=True)
    print(f"Created output directory: {output_dir}")
    return output_dir

# Configure matplotlib to prevent memory warnings
plt.rcParams['figure.max_open_warning'] = 50
plt.ioff()  # Turn off interactive mode

# ===== Load and Prepare Data =====
json_path = "./Data/data2_scaled_custom_new.json"

with open(json_path, "r") as f:
    raw_data = json.load(f)

df = pd.DataFrame(raw_data)
X = df.drop(columns=["Creep rupture life"]).values.astype(np.float32)
y = df["Creep rupture life"].values.astype(np.float32).reshape(-1, 1)

class CustomDataset(Dataset):
    def __init__(self, features, targets):
        self.X = torch.tensor(features)
        self.y = torch.tensor(targets)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = CustomDataset(X, y)

# ===== KORREKTE DATENAUFTEILUNG: Train-Validation-Test Split =====
from sklearn.model_selection import train_test_split

def create_proper_data_splits(dataset, test_size=0.2, val_size=0.2, random_state=42):
    """
    Erstellt korrekte Train-Validation-Test Aufteilung
    
    Args:
        dataset: PyTorch Dataset
        test_size: Anteil Test-Set (default: 15%)
        val_size: Anteil Validation-Set (default: 15%) 
        random_state: Für Reproduzierbarkeit
        
    Returns:
        train_indices, val_indices, test_indices
    """
    
    # Gesamte Indizes
    all_indices = list(range(len(dataset)))
    
    # Erster Split: Train+Val vs Test
    train_val_indices, test_indices = train_test_split(
        all_indices, 
        test_size=test_size, 
        random_state=random_state,
        shuffle=True
    )
    
    # Zweiter Split: Train vs Validation
    # val_size bezogen auf ursprüngliche Datensatz-Größe
    val_size_adjusted = val_size / (1 - test_size)
    
    train_indices, val_indices = train_test_split(
        train_val_indices,
        test_size=val_size_adjusted,
        random_state=random_state,
        shuffle=True
    )
    
    print(f"✅ Korrekte Datenaufteilung:")
    print(f"  Training:   {len(train_indices):4d} Samples ({len(train_indices)/len(dataset)*100:.1f}%)")
    print(f"  Validation: {len(val_indices):4d} Samples ({len(val_indices)/len(dataset)*100:.1f}%)")
    print(f"  Test:       {len(test_indices):4d} Samples ({len(test_indices)/len(dataset)*100:.1f}%)")
    print(f"  Total:      {len(dataset):4d} Samples")
    print(f"🔒 Test-Set wird NIEMALS für Training oder Hyperparameter-Optimierung verwendet!")
    
    return train_indices, val_indices, test_indices

# Erstelle korrekte Aufteilung
train_indices, val_indices, test_indices = create_proper_data_splits(dataset)

# Erstelle Subsets
train_dataset = Subset(dataset, train_indices)
val_dataset = Subset(dataset, val_indices) 
test_dataset = Subset(dataset, test_indices)

# ===== Enhanced Neural Network Model with Flexible Architecture =====
class RegressionNN(nn.Module):
    def __init__(self, input_dim, hidden_layers, dropout_rates, activation='ReLU'):
        super(RegressionNN, self).__init__()
        
        # Choose activation function
        if activation == 'ReLU':
            act_fn = nn.ReLU
        elif activation == 'LeakyReLU':
            act_fn = nn.LeakyReLU
        elif activation == 'ELU':
            act_fn = nn.ELU
        elif activation == 'GELU':
            act_fn = nn.GELU
        else:
            act_fn = nn.ReLU
        
        layers = []
        prev_dim = input_dim
        
        for i, (hidden_dim, dropout_rate) in enumerate(zip(hidden_layers, dropout_rates)):
            layers.append(nn.Linear(prev_dim, hidden_dim))
            layers.append(act_fn())
            if dropout_rate > 0:
                layers.append(nn.Dropout(dropout_rate))
            prev_dim = hidden_dim
        
        # Output layer
        layers.append(nn.Linear(prev_dim, 1))
        
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

# ===== Training Function with Enhanced Features (R² Optimization) =====
def train_model(model, train_loader, val_loader, trial_params, max_epochs=3000, return_history=False):
    criterion = nn.MSELoss()
    
    # Optimizer selection
    if trial_params['optimizer'] == 'Adam':
        optimizer = optim.Adam(
            model.parameters(), 
            lr=trial_params['lr'], 
            weight_decay=trial_params['weight_decay']
        )
    elif trial_params['optimizer'] == 'AdamW':
        optimizer = optim.AdamW(
            model.parameters(), 
            lr=trial_params['lr'], 
            weight_decay=trial_params['weight_decay']
        )
    else:  # SGD
        optimizer = optim.SGD(
            model.parameters(), 
            lr=trial_params['lr'], 
            weight_decay=trial_params['weight_decay'],
            momentum=trial_params.get('momentum', 0.9)
        )
    
    # Scheduler (adjusted for R² maximization)
    if trial_params['scheduler'] == 'ReduceLROnPlateau':
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='max',  # Changed to 'max' for R² maximization
            factor=trial_params['scheduler_factor'], 
            patience=trial_params['scheduler_patience']
        )
    elif trial_params['scheduler'] == 'CosineAnnealingLR':
        scheduler = optim.lr_scheduler.CosineAnnealingLR(
            optimizer, T_max=max_epochs//4
        )
    else:  # StepLR
        scheduler = optim.lr_scheduler.StepLR(
            optimizer, 
            step_size=trial_params['step_size'], 
            gamma=trial_params['gamma']
        )
    
    model.train()
    train_losses, val_losses, val_r2_scores = [], [], []
    
    # Early stopping parameters (now based on R²)
    early_stopping_patience = trial_params['early_stopping_patience']
    early_stopping_counter = 0
    best_val_r2 = -float('inf')  # Changed to maximize R²
    best_model_state = None

    for epoch in range(max_epochs):
        epoch_train_loss = 0
        for xb, yb in train_loader:
            optimizer.zero_grad()
            pred = model(xb)
            loss = criterion(pred, yb)
            loss.backward()
            
            # Gradient clipping
            if trial_params['grad_clip'] > 0:
                torch.nn.utils.clip_grad_norm_(model.parameters(), trial_params['grad_clip'])
            
            optimizer.step()
            epoch_train_loss += loss.item()

        train_losses.append(epoch_train_loss / len(train_loader))

        # Evaluate on validation set (calculate both loss and R²)
        model.eval()
        val_loss = 0
        val_preds, val_targets = [], []
        
        with torch.no_grad():
            for xb, yb in val_loader:
                pred = model(xb)
                loss = criterion(pred, yb)
                val_loss += loss.item()
                val_preds.extend(pred.numpy().flatten())
                val_targets.extend(yb.numpy().flatten())

        val_loss /= len(val_loader)
        val_losses.append(val_loss)
        
        # Calculate R² score
        val_r2 = r2_score(val_targets, val_preds)
        val_r2_scores.append(val_r2)

        # Learning rate scheduling (use R² for ReduceLROnPlateau)
        if trial_params['scheduler'] == 'ReduceLROnPlateau':
            scheduler.step(val_r2)  # Use R² instead of loss
        else:
            scheduler.step()

        # Early stopping check (based on R²)
        if val_r2 > best_val_r2:
            best_val_r2 = val_r2
            early_stopping_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            early_stopping_counter += 1

        # Early stopping
        if early_stopping_counter >= early_stopping_patience:
            break

        model.train()
    
    # Load best model
    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    if return_history:
        return best_val_r2, {'train_losses': train_losses, 'val_losses': val_losses, 'val_r2_scores': val_r2_scores}
    else:
        return best_val_r2  # Return R² instead of loss

# ===== Evaluation =====
def evaluate_model(model, loader):
    model.eval()
    preds, targets = [], []
    with torch.no_grad():
        for xb, yb in loader:
            pred = model(xb)
            preds.append(pred.numpy())
            targets.append(yb.numpy())
    preds = np.vstack(preds)
    targets = np.vstack(targets)
    return preds, targets

# ===== Objective Function for Optuna (R² Maximization) - KORRIGIERT =====
def objective(trial):
    """
    🔥 KORRIGIERTE Objective Function - verwendet NUR Train+Validation, 
    NIEMALS das Test-Set!
    """
    # Hyperparameters to optimize
    
    # Architecture parameters
    n_layers = trial.suggest_int('n_layers', 2, 3)
    hidden_layers = []
    dropout_rates = []
    
    for i in range(n_layers):
        # Layer size - decreasing pattern
        if i == 0:
            layer_size = trial.suggest_int(f'layer_{i}_size', 64, 256, step=16)
        else:
            prev_size = hidden_layers[i-1]
            layer_size = trial.suggest_int(f'layer_{i}_size', 32, min(prev_size, 256), step=16)
        
        hidden_layers.append(layer_size)
        dropout_rates.append(trial.suggest_float(f'dropout_{i}', 0.1, 0.5, step=0.05))
    
    # Training parameters
    trial_params = {
        'lr': trial.suggest_float('lr', 1e-5, 1e-2, log=True),
        'weight_decay': trial.suggest_float('weight_decay', 1e-6, 1e-2, log=True),
        'batch_size': trial.suggest_categorical('batch_size', [4, 8, 12, 16, 32,]),
        'optimizer': trial.suggest_categorical('optimizer', ['Adam', 'AdamW', 'SGD']),
        'activation': trial.suggest_categorical('activation', ['ReLU', 'LeakyReLU', 'ELU', 'GELU']),
        'scheduler': trial.suggest_categorical('scheduler', ['ReduceLROnPlateau', 'CosineAnnealingLR', 'StepLR']),
        'early_stopping_patience': trial.suggest_int('early_stopping_patience', 20, 40),
        'grad_clip': trial.suggest_float('grad_clip', 0.0, 2.0, step=0.5)
    }
    
    # Scheduler-specific parameters
    if trial_params['scheduler'] == 'ReduceLROnPlateau':
        trial_params['scheduler_factor'] = trial.suggest_float('scheduler_factor', 0.1, 0.7, step=0.1)
        trial_params['scheduler_patience'] = trial.suggest_int('scheduler_patience', 3, 10)
    elif trial_params['scheduler'] == 'StepLR':
        trial_params['step_size'] = trial.suggest_int('step_size', 50, 200, step=25)
        trial_params['gamma'] = trial.suggest_float('gamma', 0.1, 0.9, step=0.1)
    
    # SGD-specific parameters
    if trial_params['optimizer'] == 'SGD':
        trial_params['momentum'] = trial.suggest_float('momentum', 0.8, 0.99, step=0.05)
    
    # 🔥 WICHTIG: Verwende nur TRAIN+VALIDATION für CV, NICHT das Test-Set!
    k = 5  # 3-Fold CV nur auf Train+Val Daten
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    cv_r2_scores = []
    
    # Kombiniere train und validation für CV (Test-Set bleibt unberührt!)
    train_val_combined = train_indices + val_indices
    train_val_dataset = Subset(dataset, train_val_combined)
    
    for fold, (fold_train_idx, fold_val_idx) in enumerate(kf.split(train_val_combined)):
        # Erstelle Subsets für diesen Fold
        fold_train_subset = Subset(train_val_dataset, fold_train_idx)
        fold_val_subset = Subset(train_val_dataset, fold_val_idx)
        
        train_loader = DataLoader(fold_train_subset, batch_size=trial_params['batch_size'], shuffle=True)
        val_loader = DataLoader(fold_val_subset, batch_size=trial_params['batch_size'])
        
        model = RegressionNN(
            input_dim=20, 
            hidden_layers=hidden_layers, 
            dropout_rates=dropout_rates,
            activation=trial_params['activation']
        )
        
        val_r2 = train_model(model, train_loader, val_loader, trial_params, max_epochs=250)
        cv_r2_scores.append(val_r2)
        
        # Pruning unpromising trials (inverted logic for R² maximization)
        if fold == 0:  # Prune after first fold if R² is poor
            trial.report(val_r2, fold)
            if trial.should_prune():
                raise optuna.exceptions.TrialPruned()
    
    return np.mean(cv_r2_scores)  # Return mean R² score

# ===== Run Optuna Optimization (R² Maximization) =====
def run_optimization(n_trials=50, n_jobs=4):
    print("Starting Optuna hyperparameter optimization for R² maximization...")
    print(f"Running {n_trials} trials with {n_jobs} parallel job(s)")
    
    # Create study (changed direction to 'maximize' for R²)
    study = optuna.create_study(
        direction='maximize',  # Changed to maximize R²
        sampler=optuna.samplers.TPESampler(seed=42),
        pruner=optuna.pruners.MedianPruner(
            n_startup_trials=10, 
            n_warmup_steps=5,
            n_min_trials=5  # Added for R² optimization
        )
    )
    
    # Optimize
    study.optimize(objective, n_trials=n_trials, n_jobs=n_jobs)
    
    print("\nOptimization completed!")
    print(f"Number of finished trials: {len(study.trials)}")
    print(f"Number of pruned trials: {len([t for t in study.trials if t.state == TrialState.PRUNED])}")
    print(f"Number of complete trials: {len([t for t in study.trials if t.state == TrialState.COMPLETE])}")
    
    # Best trial #auf welchen daten
    best_trial = study.best_trial
    print(f"\nBest trial R²: {best_trial.value:.6f}")
    print("Best parameters:")
    for key, value in best_trial.params.items():
        print(f"  {key}: {value}")
    
    return study, best_trial

# ===== Train Final Model with Best Parameters (R² Optimized) - KORRIGIERT =====
def train_final_model(best_params, save_path="./NN/final_models"):
    """
    🔥 KORRIGIERTES finales Training NUR auf Train+Validation Daten.
    Test-Set bleibt für finale Evaluation unberührt!
    """
    print("\n🔥 Training final model OHNE Test-Set (korrekte Methodik)...")
    
    # Extract architecture parameters
    n_layers = best_params['n_layers']
    hidden_layers = [best_params[f'layer_{i}_size'] for i in range(n_layers)]
    dropout_rates = [best_params[f'dropout_{i}'] for i in range(n_layers)]
    
    # Prepare trial_params dict
    trial_params = {k: v for k, v in best_params.items() 
                   if not k.startswith('layer_') and not k.startswith('dropout_') and k != 'n_layers'}
    
    # 🔥 WICHTIG: Finale Evaluation nur auf Train+Validation!
    print("🔒 Test-Set wird NICHT für Training verwendet!")
    
    # Train final model auf alle Train+Val Daten
    train_val_combined = train_indices + val_indices
    train_val_dataset = Subset(dataset, train_val_combined)
    
    # Split für finale Evaluation: 80% train, 20% val aus Train+Val Daten
    final_train_idx, final_val_idx = train_test_split(
        range(len(train_val_dataset)), 
        test_size=0.2, 
        random_state=42
    )
    
    final_train_subset = Subset(train_val_dataset, final_train_idx)
    final_val_subset = Subset(train_val_dataset, final_val_idx)
    
    train_loader = DataLoader(final_train_subset, batch_size=best_params['batch_size'], shuffle=True)
    val_loader = DataLoader(final_val_subset, batch_size=best_params['batch_size'])
    
    # Create and train final model
    final_model = RegressionNN(
        input_dim=20, 
        hidden_layers=hidden_layers, 
        dropout_rates=dropout_rates,
        activation=best_params['activation']
    )
    
    print(f"Training auf {len(final_train_subset)} Samples, Validierung auf {len(final_val_subset)} Samples")
    
    # Train mit Geschichte
    final_r2, training_history = train_model(
        final_model, train_loader, val_loader, trial_params, 
        max_epochs=5000, return_history=True
    )
    
    # Evaluate auf validation portion
    preds, targets = evaluate_model(final_model, val_loader)
    mse = mean_squared_error(targets, preds)
    r2 = r2_score(targets, preds)
    
    print(f"\n✅ Final Model Performance (auf Validation aus Train+Val):")
    print(f"   R²: {r2:.6f}")
    print(f"   MSE: {mse:.6f}")
    print(f"   RMSE: {np.sqrt(mse):.6f}")
    
    # Save model
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    model_path = os.path.join(save_path, f"corrected_model_{timestamp}.pth")
    torch.save(final_model.state_dict(), model_path)
    
    params_path = os.path.join(save_path, f"corrected_params_{timestamp}.json")
    with open(params_path, 'w') as f:
        json.dump(best_params, f, indent=2)
    
    print(f"\n💾 Model saved: {model_path}")
    print(f"💾 Parameters saved: {params_path}")
    
    return final_model, [(mse, r2)], training_history

# ===== Enhanced Visualization Functions =====
def plot_parity_plot(model, data_loader, title="Parity Plot", output_dir=None, save_prefix="parity_plot"):
    """Create a simple parity plot with R² and MSE metrics"""
    preds, targets = evaluate_model(model, data_loader)
    
    # Flatten arrays
    preds = preds.flatten()
    targets = targets.flatten()
    
    # Calculate metrics
    r2 = r2_score(targets, preds)
    mse = mean_squared_error(targets, preds)
    rmse = np.sqrt(mse)
    
    # Create figure
    plt.figure(figsize=(8, 8))
    
    # Create scatter plot
    plt.scatter(targets, preds, alpha=0.7, s=30, c='blue', edgecolors='none')
    
    # Add perfect prediction line (y=x)
    min_val = min(min(targets), min(preds))
    max_val = max(max(targets), max(preds))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction (y=x)')
    
    # Formatting
    plt.xlabel('True Values', fontsize=12, fontweight='bold')
    plt.ylabel('Predicted Values', fontsize=12, fontweight='bold')
    plt.title(f'{title}\nR² = {r2:.4f}, MSE = {mse:.4f}, RMSE = {rmse:.4f}', fontsize=14, fontweight='bold')
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)
    
    # Equal aspect ratio
    plt.gca().set_aspect('equal', adjustable='box')
    plt.tight_layout()
    
    # Save plot if output directory is provided
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"{save_prefix}_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    return r2, mse, rmse

def plot_training_history(training_history, title="Training History", output_dir=None, save_prefix="training_history"):
    """Plot training and validation loss/R² over epochs"""
    if training_history is None:
        print("No training history available. Make sure to call train_model with return_history=True")
        return
    
    train_losses = training_history['train_losses']
    val_losses = training_history['val_losses']
    val_r2_scores = training_history['val_r2_scores']
    
    epochs = range(1, len(train_losses) + 1)
    
    # Plot 1: Loss curves
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_losses, 'b-', label='Training Loss', linewidth=2)
    plt.plot(epochs, val_losses, 'r-', label='Validation Loss', linewidth=2)
    plt.xlabel('Epochs', fontweight='bold')
    plt.ylabel('Loss (MSE)', fontweight='bold')
    plt.title('Training and Validation Loss', fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.yscale('log')
    plt.tight_layout()
    
    # Save loss plot
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"training_loss_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    # Plot 2: R² scores
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, val_r2_scores, 'g-', label='Validation R²', linewidth=2)
    plt.xlabel('Epochs', fontweight='bold')
    plt.ylabel('R² Score', fontweight='bold')
    plt.title('Validation R² Score', fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.ylim([0, 1])
    plt.tight_layout()
    
    # Save R² plot
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"training_r2_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()

def plot_residuals(model, data_loader, title="Residual Plot", output_dir=None, save_prefix="residuals"):
    """Create residual plots as individual plots"""
    preds, targets = evaluate_model(model, data_loader)
    
    # Flatten arrays
    preds = preds.flatten()
    targets = targets.flatten()
    
    # Calculate residuals
    residuals = targets - preds
    
    # 1. Residuals vs Predicted Values
    plt.figure(figsize=(8, 6))
    plt.scatter(preds, residuals, alpha=0.7, s=30, c='blue', edgecolors='none')
    plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
    plt.xlabel('Predicted Values', fontweight='bold')
    plt.ylabel('Residuals (True - Predicted)', fontweight='bold')
    plt.title('Residuals vs Predicted Values', fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"residuals_vs_predicted_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    # 2. Residuals vs True Values
    plt.figure(figsize=(8, 6))
    plt.scatter(targets, residuals, alpha=0.7, s=30, c='green', edgecolors='none')
    plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
    plt.xlabel('True Values', fontweight='bold')
    plt.ylabel('Residuals (True - Predicted)', fontweight='bold')
    plt.title('Residuals vs True Values', fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"residuals_vs_true_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    # 3. Histogram of Residuals
    plt.figure(figsize=(8, 6))
    plt.hist(residuals, bins=30, alpha=0.7, color='orange', edgecolor='black')
    plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
    plt.xlabel('Residuals', fontweight='bold')
    plt.ylabel('Frequency', fontweight='bold')
    plt.title(f'Distribution of Residuals\nMean: {np.mean(residuals):.4f}, Std: {np.std(residuals):.4f}', fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"residuals_histogram_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    # 4. Q-Q Plot (Normal distribution check)
    plt.figure(figsize=(8, 6))
    from scipy import stats
    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title('Q-Q Plot (Normal Distribution Check)', fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.gca().get_lines()[0].set_markerfacecolor('purple')
    plt.gca().get_lines()[0].set_markeredgecolor('purple')
    plt.gca().get_lines()[1].set_color('red')
    plt.tight_layout()
    
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"residuals_qq_plot_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    # Print some statistics
    print(f"Residual Statistics:")
    print(f"Mean: {np.mean(residuals):.6f}")
    print(f"Standard Deviation: {np.std(residuals):.6f}")
    print(f"Min: {np.min(residuals):.6f}")
    print(f"Max: {np.max(residuals):.6f}")
    print(f"Mean Absolute Error: {np.mean(np.abs(residuals)):.6f}")

def plot_comprehensive_model_analysis(model, dataset, best_params, training_history=None, output_dir=None):
    """Comprehensive analysis of the best model with all plots"""
    print("Generating comprehensive model analysis...")
    
    # 🔥 KORRIGIERT: Verwende die vordefinierten Train/Val/Test Splits!
    # Nicht mehr random splits, sondern die bereits erstellten!
    
    # Create data loaders für alle drei Sets
    full_loader = DataLoader(dataset, batch_size=64, shuffle=False)
    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=False)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
    
    print("\n" + "="*60)
    print("1. PARITY PLOT (Full Dataset)")
    print("="*60)
    r2_full, mse_full, rmse_full = plot_parity_plot(
        model, full_loader, "Parity Plot - Full Dataset", output_dir, "parity_plot_full"
    )
    
    print("\n" + "="*60)
    print("2. PARITY PLOT (Validation Set - aus Train+Val)")
    print("="*60)
    r2_val, mse_val, rmse_val = plot_parity_plot(
        model, val_loader, "Parity Plot - Validation Set", output_dir, "parity_plot_val"
    )
    
    if training_history is not None:
        print("\n" + "="*60)
        print("3. TRAINING HISTORY")
        print("="*60)
        plot_training_history(
            training_history, "Training and Validation Error Over Epochs", output_dir, "training_history"
        )
    
    print("\n" + "="*60)
    print("4. RESIDUAL ANALYSIS (Validation Set)")
    print("="*60)
    plot_residuals(model, val_loader, "Residual Analysis - Validation Set", output_dir, "residuals_val")
    
    # Save model summary to text file
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        summary_file = os.path.join(output_dir, f"model_summary_{timestamp}.txt")
        with open(summary_file, 'w') as f:
            f.write("MODEL SUMMARY (CORRECTED METHODOLOGY)\n")
            f.write("="*60 + "\n")
            f.write(f"Architecture: {[best_params[f'layer_{i}_size'] for i in range(best_params['n_layers'])]}\n")
            f.write(f"Activation: {best_params['activation']}\n")
            f.write(f"Optimizer: {best_params['optimizer']}\n")
            f.write(f"Learning Rate: {best_params['lr']:.2e}\n")
            f.write(f"Batch Size: {best_params['batch_size']}\n")
            f.write("\nData Split:\n")
            f.write(f"Training: {len(train_dataset)} samples\n")
            f.write(f"Validation: {len(val_dataset)} samples\n")
            f.write(f"Test: {len(test_dataset)} samples (UNUSED for training/optimization)\n")
            f.write("\nPerformance Metrics:\n")
            f.write(f"Full Dataset - R²: {r2_full:.6f}, MSE: {mse_full:.6f}, RMSE: {rmse_full:.6f}\n")
            f.write(f"Validation Set - R²: {r2_val:.6f}, MSE: {mse_val:.6f}, RMSE: {rmse_val:.6f}\n")
            f.write("\nNOTE: Test set performance will be evaluated separately!\n")
        print(f"Model summary saved: {summary_file}")
    
    # Print model summary
    print("\n" + "="*60)
    print("MODEL SUMMARY (CORRECTED METHODOLOGY)")
    print("="*60)
    print(f"Architecture: {[best_params[f'layer_{i}_size'] for i in range(best_params['n_layers'])]}")
    print(f"Activation: {best_params['activation']}")
    print(f"Optimizer: {best_params['optimizer']}")
    print(f"Learning Rate: {best_params['lr']:.2e}")
    print(f"Batch Size: {best_params['batch_size']}")
    print(f"\nData Split:")
    print(f"Training: {len(train_dataset)} samples")
    print(f"Validation: {len(val_dataset)} samples") 
    print(f"Test: {len(test_dataset)} samples (🔒 UNUSED for training/optimization)")
    print("\nPerformance Metrics:")
    print(f"Full Dataset - R²: {r2_full:.6f}, MSE: {mse_full:.6f}, RMSE: {rmse_full:.6f}")
    print(f"Validation Set - R²: {r2_val:.6f}, MSE: {mse_val:.6f}, RMSE: {rmse_val:.6f}")
    print("\n🎯 NOTE: Test set performance will be evaluated separately!")

# ===== Visualization Functions =====
def plot_optimization_history(study, output_dir=None, save_prefix="optimization_history"):
    """Plot optimization history as separate plots"""
    # Plot 1: Optimization History
    plt.figure(figsize=(10, 6))
    optuna.visualization.matplotlib.plot_optimization_history(study)
    plt.title("Optimization History", fontweight='bold', fontsize=14)
    plt.xlabel("Trial", fontweight='bold')
    plt.ylabel("R² Score", fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.ylim(0, 1)  # Set y-axis from 0 to 1
    
    # Save optimization history plot
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"optimization_history_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()
    
    # Plot 2: Parameter Importances
    plt.figure(figsize=(10, 6))
    optuna.visualization.matplotlib.plot_param_importances(study)
    plt.title("Parameter Importances", fontweight='bold', fontsize=14)
    plt.xlabel("Importance", fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    # Save parameter importances plot
    if output_dir is not None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"parameter_importances_{timestamp}.png"
        filepath = os.path.join(output_dir, filename)
        plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Plot saved: {filepath}")
    
    plt.show()
    plt.close()

def plot_hyperparameter_relationships(study, output_dir=None, save_prefix="hyperparameter_relationships"):
    """Plot hyperparameter relationships as individual plots"""
    # Get important parameters
    importances = optuna.importance.get_param_importances(study)
    top_params = list(importances.keys())[:6]  # Top 6 parameters
    
    for i, param in enumerate(top_params):
        plt.figure(figsize=(8, 6))
        
        # Get values and objectives
        values = []
        objectives = []
        for trial in study.trials:
            if trial.state == TrialState.COMPLETE and param in trial.params:
                values.append(trial.params[param])
                objectives.append(trial.value)
        
        plt.scatter(values, objectives, alpha=0.7, s=50, c='blue', edgecolors='none')
        plt.xlabel(param, fontweight='bold', fontsize=12)
        plt.ylabel('R² Score', fontweight='bold', fontsize=12)
        plt.title(f'{param} vs R²', fontweight='bold', fontsize=14)
        plt.ylim(0, 1)  # Set y-axis from 0 to 1
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        
        # Save individual plot
        if output_dir is not None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = f"hyperparameter_{param}_{timestamp}.png"
            filepath = os.path.join(output_dir, filename)
            plt.savefig(filepath, dpi=300, bbox_inches='tight', facecolor='white')
            print(f"Plot saved: {filepath}")
        
        plt.show()
        plt.close()

# ===== FINALE TEST-SET EVALUATION (Die Wahrheit!) =====
def evaluate_on_test_set(model, output_dir=None):
    """
    🎯 FINALE EVALUATION auf dem unberührten Test-Set!
    Das ist die wahre Performance des Modells!
    """
    print("\n" + "="*70)
    print("🎯 FINALE TEST-SET EVALUATION - Die Moment der Wahrheit!")
    print("="*70)
    
    # Create test loader
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
    
    # Evaluate
    preds, targets = evaluate_model(model, test_loader)
    
    # Calculate metrics
    mse = mean_squared_error(targets, preds)
    r2 = r2_score(targets, preds)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(targets - preds))
    
    print(f"📊 FINALE TEST-SET PERFORMANCE:")
    print(f"   R² Score:  {r2:.6f}")
    print(f"   MSE:       {mse:.6f}")
    print(f"   RMSE:      {rmse:.6f}")
    print(f"   MAE:       {mae:.6f}")
    print(f"   Samples:   {len(test_dataset)}")
    
    # Performance Interpretation
    print(f"\n🎭 Performance Interpretation:")
    if r2 > 0.9:
        print("   🟢 Excellent performance (R² > 0.9)")
    elif r2 > 0.8:
        print("   🟡 Good performance (R² > 0.8)")
    elif r2 > 0.6:
        print("   🟠 Moderate performance (R² > 0.6)")
    else:
        print("   🔴 Poor performance (R² < 0.6)")
    
    # Erstelle Test-Set Parity Plot
    plt.figure(figsize=(8, 8))
    plt.scatter(targets.flatten(), preds.flatten(), alpha=0.7, s=50, c='red', edgecolors='black')
    
    min_val = min(min(targets.flatten()), min(preds.flatten()))
    max_val = max(max(targets.flatten()), max(preds.flatten()))
    plt.plot([min_val, max_val], [min_val, max_val], 'k--', linewidth=3, label='Perfect Prediction')
    
    plt.xlabel('True Values (Test Set)', fontsize=14, fontweight='bold')
    plt.ylabel('Predicted Values (Test Set)', fontsize=14, fontweight='bold')
    plt.title(f'🎯 FINAL TEST-SET EVALUATION\\nR² = {r2:.4f}, RMSE = {rmse:.4f}', 
              fontsize=16, fontweight='bold')
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.gca().set_aspect('equal', adjustable='box')
    
    # Save plot
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    test_plot_path = f"./FINAL_TEST_EVALUATION_{timestamp}.png"
    plt.savefig(test_plot_path, dpi=300, bbox_inches='tight', facecolor='white')
    
    plt.tight_layout()
    plt.show()
    plt.close()
    
    print(f"\n🏆 DIESE WERTE REPRÄSENTIEREN DIE WAHRE MODEL-PERFORMANCE!")
    print(f"📈 Plot gespeichert als: FINAL_TEST_EVALUATION_{timestamp}.png")
    
    return r2, mse, rmse, mae

# ===== Main Execution - KORRIGIERTE VERSION =====
if __name__ == "__main__":
    print("🚀 Starte KORRIGIERTE ML-Pipeline mit Train-Val-Test Split...")
    print("=" * 60)
    
    # Run hyperparameter optimization (nur auf Train+Val)
    print("⚡ Phase 1: Hyperparameter Optimization (Train+Val only)")
    study, best_trial = run_optimization(n_trials=50, n_jobs=4)  # Reduziert für Demo
    output_dir = create_output_directory()    
    
    # Plot optimization results
    print("\n⚡ Phase 2: Plotting Optimization Results")
    plot_optimization_history(study, output_dir, "optimization_history")
    plot_hyperparameter_relationships(study, output_dir, "hyperparameter_relationships")
    
    # Train final model with best parameters (nur auf Train+Val)
    print("\n⚡ Phase 3: Final Model Training (Train+Val only)")
    best_model, final_results, training_history = train_final_model(best_trial.params)
    
    # Comprehensive model analysis (ohne Test-Set)
    print("\n⚡ Phase 4: Model Analysis (ohne Test-Set)")
    plot_comprehensive_model_analysis(best_model, dataset, best_trial.params, training_history, output_dir)
    
    # 🎯 FINALE TEST-SET EVALUATION (die Wahrheit!)
    print("\n⚡ Phase 5: FINAL TEST-SET EVALUATION (Moment der Wahrheit!)")
    test_r2, test_mse, test_rmse, test_mae = evaluate_on_test_set(best_model, output_dir)
    
    # Vergleiche Validation vs Test Performance
    val_r2 = final_results[0][1]  # R² from validation
    print("\n" + "="*70)
    print("📊 VALIDATION vs TEST PERFORMANCE VERGLEICH:")
    print("="*70)
    print(f"Validation R² (Train+Val): {val_r2:.6f}")
    print(f"Test R² (ungesehen):       {test_r2:.6f}")
    print(f"Differenz:                 {abs(val_r2 - test_r2):.6f}")
    
    if abs(val_r2 - test_r2) < 0.05:
        print("✅ Ausgezeichnete Generalisierung! (Differenz < 0.05)")
    elif abs(val_r2 - test_r2) < 0.10:
        print("⚠️  Moderate Generalisierung (Differenz < 0.10)")
    else:
        print("🚨 Schlechte Generalisierung! Mögliches Overfitting!")
    
    # Save study for future reference
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    study_path = f"./NN/corrected_optuna_study_{timestamp}.pkl"
    with open(study_path, 'wb') as f:
        pickle.dump(study, f)
    print(f"\nOptuna study saved to: {study_path}")
    
    print(f"\n🏆 KORRIGIERTE R² OPTIMIZATION UND FINAL TRAINING ABGESCHLOSSEN!")
    print(f"📂 All plots saved in: {output_dir}")
    print(f"🎯 TRUE TEST PERFORMANCE: R² = {test_r2:.6f}")
    print("\n🔥 METHODISCH SAUBERE EVALUATION DURCHGEFÜHRT! 🔥")
```
